# Welcome to Jupyter!

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df =pd.read_csv("area.csv")
```

```python
%matplotlib inline
plt.title('Home Price Prediction For Given Area')
font1 = {'family':'fantasy','color':'blue','size':20}
plt.xlabel('area',fontdict = font1)
plt.ylabel('price',fontdict = font1)
plt.plot(df.area,df.price, color='green', linestyle='dashed', linewidth = 2,
         marker='o', markerfacecolor='blue', markersize=10)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
d=pd.read_csv('test.csv')
```

```python
print(d)
```

```python
p=reg.predict(d)
```

```python
d['price']=p
```

```python
print(d)
```

```python
%matplotlib inline
plt.title('Home Price Prediction For Given Area')
font1 = {'family':'fantasy','color':'blue','size':20}
plt.xlabel('area',fontdict = font1)
plt.ylabel('price',fontdict = font1)
plt.plot(d.area,d.price, color='green', linestyle='dashed', linewidth = 2,
         marker='o', markerfacecolor='blue', markersize=10)
```
